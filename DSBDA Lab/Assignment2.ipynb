```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
df=pd.read_csv("StudentsPerformance.csv")
df.isnull()
```

```python
series = pd.isnull(df["Math_Score"])
df[series]
```

```python
df.notnull()
```

```python
series = pd.notnull(df["Math_Score"])
df[series]
```

```python
ndf=df
ndf.fillna(0)
```

```python
m_v=df['Math_Score'].mean()
df['Math_Score'].fillna(value=m_v, inplace=True)
df
```

```python
df.replace(to_replace = np.nan, value = -99)
```

```python
df.dropna()
```

```python
df.dropna(axis = 1)
```

```python
col = ['Math_Score', 'Reading_Score' , 'Writing_Score','Placement_Score']
df.boxplot(col)
```

```python
df['Math_Score'].plot(kind ='hist')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.title('Histogram of Math Scores')
plt.show()
```

```python
df['log_math'] = np.log10(df['Math_Score'])
df['log_math'].plot(kind = 'hist')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.title('Histogram of Math Scores')
plt.show()
```
