```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt

a = sb.load_dataset("titanic")
a.head(5)
```

```python
sb.distplot(x = a['age'], bins = 10)
```

```python
sb.distplot(a['age'], bins = 10,kde=False)
```

```python
sb.jointplot(x = a['age'], y = a['fare'], kind =
'scatter')
```

```python
sb.jointplot(x = a['age'], y = a['fare'], kind = 'hex')
```

```python
sb.rugplot(a['fare'])
```

```python
sb.barplot(x='sex', y='age', data=a, estimator=np.std)
```
